```python
review1 = pd.read_csv(r'C:\Users\Garrett Pinkston\Desktop\TIM147\Project\data\RestaurantReview1.csv')
```

```python
review1.head(3)
```

```python
reviews = review1.sort_values(by='date')
reviews = reviews.groupby('business_id').agg(list).reset_index()
reviews.drop_duplicates(subset='business_id',inplace=True)
reviews.head(3)
```

```python
review1['business_id'].nunique()
```

```python
reviews = reviews.rename(columns={'stars': 'starList','date':'reviewDate'})
reviews = reviews[['business_id','starList','useful','funny','cool','reviewDate']]
final = restaurants.merge(reviews,on='business_id')
```

```python
i=0
print(len(final['starList'].iloc[i]))
print(len(final['reviewDate'].iloc[i]))
```

```python
final = final.rename(columns={'dates': 'checkinDate'})
```

```python
final.drop(['Unnamed: 0'],axis=1,inplace=True)
```
