```python
import os

import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv(r'C:\Users\Garrett Pinkston\Desktop\TIM147\Project\data\RestaurantReview1.csv')
df2 = pd.read_csv(r'C:\Users\Garrett Pinkston\Desktop\TIM147\Project\data\RestaurantReview2.csv')
```

```python
df1.drop(df1.columns[0], axis=1, inplace=True)
df2.drop(df2.columns[:2], axis=1, inplace=True)
```

```python
df1.head(3)
```

```python
df = pd.concat([df1, df2], axis=0) 
```

```python
df3 = pd.read_csv(r'C:\Users\Garrett Pinkston\Desktop\TIM147\Project\data\RestaurantInfoComplete.csv')
```

```python
df3.drop(df3.columns[:2], axis=1, inplace=True)
df3.head(3)
```

```python
finalDf = pd.merge(df3, df, on='business_id')
finalDf.head(3)
```

```python
finalDf['attributes'][1]
```

```python
openRestaurants = finalDf[finalDf['is_open']==1]
posReviewsOpenRestaurants = openRestaurants[openRestaurants['stars_y'] > 3]
```

```python
from wordcloud import WordCloud
text_data = ' '.join(posReviewsOpenRestaurants['text'].astype(str)) #join text data
```

```python
wordcloud = WordCloud(width=1200, height=600, random_state=42, max_font_size=100, background_color='white').generate(text_data)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig("WordCloud_Postive.png")
```

```python
closedRestaurants = finalDf[finalDf['is_open']==0]
negReviewsClosedRestaurants = closedRestaurants[closedRestaurants['stars_y'] < 3]
```

```python
# text of negative closed restaurants
text_data = ' '.join(negReviewsClosedRestaurants['text'].astype(str))
```

```python
#fit and display wordcloud of negative reviews for closed restaurants
wordcloud = WordCloud(width=1200, height=600, random_state=42, max_font_size=100, background_color='white').generate(text_data)


plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig("WordCloud_Negative.png")
```

```python
#get text length of each column, put in new column
finalDf['text_length'] = finalDf['text'].apply(len)
finalDf.head(3)
```

```python
#text length vs rating
plt.figure(figsize=(10, 6))
plt.scatter(finalDf['text_length'], finalDf['stars_y'], alpha=0.5)
plt.title('Text Length vs Rating')
plt.xlabel('Text Length')
plt.ylabel('Rating')
plt.show()
#since rating is discrete, this plot is useless. I'll do a histogram and boxplot to visualize rating statistics
```

```python
#histogram

plt.figure(figsize=(10, 6))
for rating in finalDf['stars_y'].unique():
    plt.hist(finalDf[finalDf['stars_y'] == rating]['text_length'], bins=30, alpha=0.5, label=f'Rating {rating}')

plt.title('Distribution of Text Length for each rating')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.legend()
plt.savefig("HistogramTextLength.png")
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='stars_y', y='text_length', data=finalDf)
plt.title('Text Length Box Plot for Each Rating')
plt.xlabel('Rating')
plt.ylabel('Text Length')
plt.savefig("BoxPlotTextLength.png")
```
